ploting and matplot 

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('problem_set_1_data.csv')
```

```python

data.head()
```

```python
type(data)
```

```python
quizzes = np.sort(data["quizzes"]) # dont ussually sort it this way
exams = np.sort(data['final_EXAM'])
x = np.arange(0,len(quizzes))
```

```python
print(x)
```

```python
# functional/procedural approach to ploting
# quick and dirty
plt.plot(x,quizzes,label = 'quizzes') # after sorting, the graph is continous 
plt.plot(x,exams, label = 'exams', linestyle = '', marker = '.', markersize = 20) # many parameters we can add

plt.xlabel('ID')
plt.ylabel('points')
plt.title("Grades in Garrett's class")

plt.legend() # show legends
plt.show()
```

```python
# subplots
plt.subplot(1,2,1)
plt.plot(x,quizzes)
plt.xlabel('ID')
plt.ylabel('points')
plt.title("Grades in Garrett's class")
plt.ylim([0,90]) 


plt.subplot(1,2,2)
plt.plot(x,exams)
plt.xlabel('ID')
plt.ylabel('points')
plt.title("Grades in Garrett's class")
plt.ylim([0,90]) # limit the range of the y axis to cope with misinterpretation
```

```python
# object oriented approach
# not as "sequential"

fig = plt.figure()
axes = fig.add_axes([.1,.1,.8,.8]) # left,bottom,width,height

axes.plot(x,quizzes)
axes.plot(x,exams)
axes.set_xlabel('ID')
axes.set_ylabel('Grades')
axes.set_title('Class__')
```

```python
fig # we can call the table
```

```python
# subplot
fig_2 = plt.figure()
axes_1 = fig_2.add_axes([0,0,.4,.8])
axes_2 = fig_2.add_axes([0.5,0,.4,.8])

axes_1.plot(x,quizzes)
axes_2.plot(x,exams)
```

```python
# using subplot to create multiple subplots

fig3, axes = plt.subplots(7,3, figsize = [10,10])

axes[0,0].plot(x,quizzes) # access to each subplot
axes[0,1].plot(x,exams)

for i in range(axes.shape[0]):
  for j in range(axes.shape[1]):
    axes[i,j].set_ylabel('Grades')
    axes[i,j].set_xlabel('ID')


plt.tight_layout()
```
