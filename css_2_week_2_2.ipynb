```python
import matplotlib.pyplot as plt 
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
tips = sns.load_dataset('tips') 
tips.head()
```

```python
tips.describe()
```

```python
# plot a single column
sns.distplot(tips['total_bill'],kde = False, bins = [0,20,40,60]) # bins cause different interpretations of the data
```

```python
sns.distplot(tips['total_bill'],kde = False, bins = range(60))
```

```python
plt.figure(figsize = [20,4])
sns.boxplot(tips['total_bill'])
plt.xticks(ticks = np.percentile(tips['total_bill'],[0,25,50,75]))
```

```python
print(sns.boxplot(y = tips['total_bill']))
```

```python
sns.boxplot(x = tips['day'], y = tips['total_bill'])
```

```python
# hue adds another dimension to data 
sns.violinplot(x = tips['day'], y = tips['total_bill'], hue = tips['sex'])
```

```python
# split works when data is dichotomous
sns.violinplot(x = tips['day'], y = tips['total_bill'], hue = tips['sex'],split = True)
```

```python
sns.violinplot(x = tips['day'], y = tips['total_bill'], hue = tips['sex'],split = True, scale = 'count')
```

```python
# scatterplot might have overlaps but stripplot creates a bit of randomness to each data
sns.stripplot(x = tips['day'], y = tips['total_bill'])
```

```python
sns.swarmplot(x = tips['day'], y = tips['total_bill'])
```

```python
# plot two columns against each other 
sns.lineplot('total_bill','tip',data = tips)
```

```python
sns.scatterplot(tips['total_bill'],tips['tip'])
```

```python
# tools in seaborn 
# does not have categorical things
sns.pairplot(tips)
```

```python
# add hue
sns.pairplot(tips,hue = 'day')
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = tips, kind = 'hex')
plt.colorbar()
```

```python
# object oriented
fig, axes = plt.subplots(2,2)
axes[0,0].plot(tips['total_bill'],tips['tip'])
sns.lineplot(tips['total_bill'],tips['tip'],ax= axes[1,1])
```
